```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)

data = pd.read_csv('foodtruck.csv')
print(data.shape)
data.head()
```

```python
X = data['Population'].values
Y = data['Profit'].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
n = len(X)
nu = 0
de = 0
for i in range (n):
    nu += (X[i] - mean_x) * (Y[i] - mean_y)
    de += (X[i] - mean_x) ** 2
m = nu/de
c = mean_y - (m * mean_x)
print(m,c,mean_y)
```

```python
# for finding the profit for jaipur city which is having population 3.073
a = float(input("enter the population of the city: "))
t = m * (a) + c
print("The profit will be: ",t*10000)
```

```python
max_x = np.max(X)
min_x = np.min(X)
x = np.linspace(min_x, max_x, 1000)
y = c + (m * x)
plt.plot(x,y, label='regression line')

plt.scatter(X,Y,label ='scatter plot')
plt.xlabel("population in 10000's")
plt.ylabel("profit in 10000 $")
plt.legend()
plt.show()
```

```python
ss_t = 0
ss_r = 0
for i in range (n):
    y_p = c + (m * x[i])
    ss_t += (Y[i] - mean_y) ** 2
    ss_r += (Y[i] - y_p) ** 2
r2 = 1 - (ss_r/ss_t)
print(r2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = X.reshape((n, 1))
reg = LinearRegression()
reg = reg.fit(X,Y)
Y_pred = reg.predict(X)
r2_score = reg.score(X,Y)
print(r2_score)
```
